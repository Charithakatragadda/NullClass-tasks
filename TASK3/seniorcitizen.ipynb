```python
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D

# Load MobileNetV2 without the top classification layer
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# Add custom layers for age and gender prediction
x = base_model.output
x = GlobalAveragePooling2D()(x)
gender_output = Dense(1, activation='sigmoid', name='gender_output')(x)  # Gender (binary classification)
age_output = Dense(1, activation='linear', name='age_output')(x)         # Age (regression)

model = Model(inputs=base_model.input, outputs=[gender_output, age_output])

# Freeze base model layers for transfer learning
for layer in base_model.layers:
    layer.trainable = False

model.compile(
    optimizer='adam',
    loss={'gender_output': 'binary_crossentropy', 'age_output': 'mse'},
    metrics={'gender_output': 'accuracy', 'age_output': 'mae'}
)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Initialize ImageDataGenerator for training and validation
datagen = ImageDataGenerator(
    rescale=1.0 / 255.0,  # Normalize pixel values to [0, 1]
    rotation_range=20,  # Augmentation: rotate images up to 20 degrees
    width_shift_range=0.2,  # Augmentation: shift images horizontally
    height_shift_range=0.2,  # Augmentation: shift images vertically
    shear_range=0.2,  # Augmentation: shear transformation
    zoom_range=0.2,  # Augmentation: zoom into images
    horizontal_flip=True,  # Augmentation: flip images horizontally
    validation_split=0.2  # Split 20% of data for validation
)

# Create training generator
train_generator = datagen.flow_from_directory(
    r"C:\Users\chari\Desktop\Age and Gender Detector\UTKFace",  # Path to your dataset
    target_size=(48,48),  # Resize images to the input size of your model
    batch_size=32,  # Number of images per batch
    class_mode='categorical',  # For multi-class classification; use 'binary' for two classes
    subset='training'  # Use the training subset
)

# Create validation generator
val_generator = datagen.flow_from_directory(
    r"C:\Users\chari\Desktop\Age and Gender Detector\UTKFace",  # Same dataset path
    target_size=(48, 48),  # Resize images to the input size of your model
    batch_size=32,  # Number of images per batch
    class_mode='categorical',  # Same as training
    subset='validation'  # Use the validation subset
)
```

```python
history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=20
)
```
